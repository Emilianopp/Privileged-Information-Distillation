```python
import torch
from transformers import AutoTokenizer

data = torch.load("/home/toolkit/outputs/traces/20260220_170640_pi_distill__tool_calls__beta0.25__alpha1.0__lr5e-06__s42/training_data/step_0001.pt", weights_only=False)
tokenizer = AutoTokenizer.from_pretrained("/mnt/adea/data_rw/finetuning/base/models--Qwen--Qwen3-8B")

print("Keys:", list(data.keys()))
for k, v in data.items():
    if isinstance(v, torch.Tensor):
        print(f"  {k}: shape={v.shape}, dtype={v.dtype}")
    else:
        print(f"  {k}: {type(v)}")
```

```python
# Pick a row to inspect
row = 0

input_ids = data["input_ids"][row]
attn_mask = data["attention_mask"][row]
responses = data["responses"][row]

# Find where the actual (non-pad) tokens start
first_real = (attn_mask == 1).nonzero(as_tuple=True)[0][0].item()
prompt_len = input_ids.shape[0] - responses.shape[0]

prompt_ids = input_ids[first_real:prompt_len]
response_ids = input_ids[prompt_len:]

print(f"Row {row}:")
print(f"  Total seq len: {input_ids.shape[0]}")
print(f"  Pad tokens: {first_real}")
print(f"  Prompt tokens: {prompt_len - first_real}")
print(f"  Response tokens: {responses.shape[0]}")
print(f"  Episode ID: {data['episode_id'][row].item()}")
print(f"  Task ID: {data['task_id'][row].item()}")
print(f"  Turn index: {data['turn_index'][row].item()}")
print(f"  Turn reward: {data['turn_reward'][row].item()}")
print(f"  Advantage: {data['advantages'][row].item()}")
print()
print("=" * 80)
print("PROMPT (decoded):")
print("=" * 80)
print(tokenizer.decode(prompt_ids, skip_special_tokens=False))
print()
print("=" * 80)
print("RESPONSE (decoded):")
print("=" * 80)
print(tokenizer.decode(response_ids, skip_special_tokens=False))
```

```python
# Compare with-PI vs without-PI prompts for pi_distill
if "input_ids_without_priv" in data:
    wp_ids = data["input_ids"][row]  # with-PI (generation prompt)
    np_ids = data["input_ids_without_priv"][row]  # without-PI
    np_mask = data["attention_mask_without_priv"][row]
    
    # Find non-pad start for without-priv
    np_first = (np_mask == 1).nonzero(as_tuple=True)[0][0].item()
    np_prompt_len = np_ids.shape[0] - responses.shape[0]
    
    wp_prompt_text = tokenizer.decode(input_ids[first_real:prompt_len], skip_special_tokens=False)
    np_prompt_text = tokenizer.decode(np_ids[np_first:np_prompt_len], skip_special_tokens=False)
    
    print(f"With-PI prompt tokens: {prompt_len - first_real}")
    print(f"Without-PI prompt tokens: {np_prompt_len - np_first}")
    print(f"Difference: {(prompt_len - first_real) - (np_prompt_len - np_first)} tokens (PI content)")
    print()
    
    # Show the diff region (system message where PI lives)
    # Find first divergence point
    wp_toks = input_ids[first_real:prompt_len].tolist()
    np_toks = np_ids[np_first:np_prompt_len].tolist()
    
    for i, (a, b) in enumerate(zip(wp_toks, np_toks)):
        if a != b:
            print(f"First divergence at token {i}")
            print(f"  With-PI:    ...{tokenizer.decode(wp_toks[max(0,i-5):i+20], skip_special_tokens=False)}...")
            print(f"  Without-PI: ...{tokenizer.decode(np_toks[max(0,i-5):i+20], skip_special_tokens=False)}...")
            break
    
    # Verify response tokens are identical
    wp_resp = input_ids[prompt_len:]
    np_resp = np_ids[np_prompt_len:]
    print(f"\nResponse tokens identical: {torch.equal(wp_resp, np_resp)}")
else:
    print("No without-PI data (not pi_distill mode)")
    
if "input_ids_with_priv" in data:
    print("\nOPSD with-PI alt tensors also present")
```

```python
# Summary table: all rows
import pandas as pd

rows = []
N = data["input_ids"].shape[0]
resp_len = data["responses"].shape[1]

for i in range(N):
    mask = data["attention_mask"][i]
    first = (mask == 1).nonzero(as_tuple=True)[0][0].item()
    plen = data["input_ids"].shape[1] - resp_len
    
    rows.append({
        "row": i,
        "episode_id": data["episode_id"][i].item(),
        "task_id": data["task_id"][i].item(),
        "turn": data["turn_index"][i].item(),
        "prompt_tokens": plen - first,
        "response_tokens": (data["responses"][i] != tokenizer.pad_token_id).sum().item(),
        "reward": data["turn_reward"][i].item(),
        "advantage": data["advantages"][i].item(),
    })

df = pd.DataFrame(rows)
print(f"Total rows: {N}")
print(f"Unique episodes: {df['episode_id'].nunique()}")
print(f"Unique tasks: {df['task_id'].nunique()}")
print()
df
```
